```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(
    n_samples=500,               # 总样本数
    centers=[(-2, 2), (2, -2)],  # **关键：类别0中心(-2,2)→左上，类别1中心(2,-2)→右下**
    cluster_std=0.5,             # 聚类标准差（值越小，点越集中）
    shuffle=True,                # 混合样本顺序
    random_state=42              # 确保结果可复现
)
```

```python
plt.figure(figsize=(7, 5))
fig, ax = plt.subplots()
ax.scatter(X[y==0][:, 0], X[y==0][:, 1], c='red',  label='Class 0', alpha=0.6)
ax.scatter(X[y==1][:, 0], X[y==1][:, 1], c='blue', label='Class 1', alpha=0.6)

xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()
diag_x = [min(xmin, ymin), max(xmax, ymax)]  # **确保覆盖整个坐标范围**

ax.plot(diag_x, diag_x, linestyle='--', color='red', label='y=x')

plt.title('Generated Binary Classification Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
